```python
df=df[df['Quantity']>0]
```

```python
df['City']=df['City'].str.strip().str.title()
```

```python
df['Category'] = df['Category'].str.strip().str.title()
```

```python
df['Total_Amount']=df['Quantity']*df['Price']
```
